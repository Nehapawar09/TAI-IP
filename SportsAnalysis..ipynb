```python
#Visualizing number of medals
ax=sns.countplot(x="Medal", width=0.3, data=df)
for p in ax.containers:
    ax.bar_label(p)
```

```python
#Number of sports
sns.countplot(y='Sport', data=df)
plt.show()
```

```python
#Number of medals won by the female and male
ax=sns.countplot(x='Gender', hue='Medal',width=0.7, data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.countplot(x='City',width=0.2, data=df)
plt.show()
```

```python
sns.countplot(x='Year', data=df, width=0.7)
plt.show()
```

```python
#Number of medal received by men and women
ax=sns.countplot(data=df, x='Gender', hue='Medal', width=0.5)
for n in ax.containers:
    ax.bar_label(n)
```
